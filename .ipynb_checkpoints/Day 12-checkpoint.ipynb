```python
def map_paths(c, current_path = ['start'], current_position = 'start'):
    for next_position in c[current_position]:
        forked_path = current_path[:]
        
        if next_position in current_path:
            if next_position == next_position.lower():
                # This means that we are going back to somewhere we can not go.
                # Path is dead
                forked_path.append('DEAD')
                continue
                
        if next_position == 'end':
            # This means that we are going to the end in a valid sequence.
            forked_path.append('VALID')
            print(forked_path)
            return forked_path
            
        # If we are here, we are not going to an invalid location, and we are not ending, so we need to go deeper
        forked_path.append(next_position)
        forked_path.append(map_paths(c, forked_path, next_position)[0])
        
    return forked_path

paths = map_paths(caverns)
```
