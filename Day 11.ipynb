```python
data = pd.read_csv('Files/Day11.txt', header=None, dtype=str)
data_v = pd.read_csv('Files/Day11Validate.txt', header=None, dtype=str)

data = data[0].str.split("", expand=True)
data = data.iloc[:,1:11]

data_v = data_v[0].str.split("", expand=True)
data_v = data_v.iloc[:,1:11]

# Question 1
# Step 1: Increase all energy by 1
# Step 2: If any of the energies is >9, flash
# Once an octopus flashes, it sets to zero and will not gain any further energy this rotation.
# Flashes increase all neighboring energies by 1

class Octopus:
    def __init__(self, energy):
        self.energy = int(energy)
        self.flash = False
        self.propogated_flash = False
        self.total_flashes = 0
    
    def increase_energy(self):
        if not self.flash:
            self.energy += 1
            if self.energy == 10:
                self.flash = True
                self.energy = 0
                self.total_flashes += 1
    
    def reset_flash(self):
        self.flash = False
        self.propogated_flash = False
            
def reset_flash(d):
    for i in range(d.shape[0]):
        for j in range(d.shape[1]):
            d.iloc[i, j].reset_flash()

def count_flashes(d):
    flashes = 0
    for i in range(d.shape[0]):
        for j in range(d.shape[1]):
            flashes += d.iloc[i, j].total_flashes
    return flashes

def display_charge(d):
    mtx = d.copy()
    for i in range(mtx.shape[0]):
        for j in range(mtx.shape[1]):
            mtx.iloc[i,j] = mtx.iloc[i,j].energy
    print(mtx)
    
def add_one_charge(d, i, j):
    d.iloc[i,j].increase_energy()
    if d.iloc[i,j].flash and not d.iloc[i,j].propogated_flash:
        # i is row
        # j is column
        d.iloc[i, j].propogated_flash = True

        left_good = j - 1 >= 0
        right_good = j + 1 < d.shape[1]
        up_good = i - 1 >= 0
        down_good = i + 1 < d.shape[0]
        
        if left_good: 
            add_one_charge(d, i, j-1)
            
        if right_good: 
            add_one_charge(d, i, j+1)
                
        if up_good: 
            add_one_charge(d, i-1, j)
            if left_good: 
                add_one_charge(d, i-1, j-1)
            if right_good:
                add_one_charge(d, i-1, j+1)
        
        if down_good: 
            add_one_charge(d, i+1, j)
            if left_good:
                add_one_charge(d, i+1, j-1)
            if right_good:
                add_one_charge(d, i+1, j+1)
                
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        data.iloc[i, j] = Octopus(data.iloc[i, j])

for i in range(100):
    display_charge(data)
    reset_flash(data)
    for j in range(data.shape[0]):
        for k in range(data.shape[1]):
            add_one_charge(data, j, k)
```

```python
# Part 2, at what point do all of the flashes sync?
def all_the_same_energy(d):
    values = []
    for i in range(d.shape[0]):
        for j in range(d.shape[1]):
            values.append(d.iloc[i,j].energy)
    unique_values = list(set(values))
    return len(unique_values) == 1

data = pd.read_csv('Files/Day11.txt', header=None, dtype=str)

data = data[0].str.split("", expand=True)
data = data.iloc[:,1:11]

for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        data.iloc[i, j] = Octopus(data.iloc[i, j])

for i in range(1000):
    reset_flash(data)
    for j in range(data.shape[0]):
        for k in range(data.shape[1]):
            add_one_charge(data, j, k)
    if all_the_same_energy(data):
        break

print(f'All the same values are at step {i+1}')
display_charge(data)
```
